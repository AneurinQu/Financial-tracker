```python
import tkinter as tk

def open_window():
    """Function to create and open a new window"""
    # Create the main window
    window = tk.Tk()
    
    # Set window properties
    window.title("Financial Tracker")
    window.geometry("800x600+300+100")  # width x height + x_offset + y_offset
    window.minsize(600, 400)
    
    # Create main frame to hold left and right sections
    main_frame = tk.Frame(window)
    main_frame.pack(fill=tk.BOTH, expand=True, padx=20, pady=20)
    
    # Left column for stored transactions
    left_frame = tk.Frame(main_frame, bg="lightgray", width=300)
    left_frame.pack(side=tk.LEFT, fill=tk.BOTH, expand=False, padx=(0, 10))
    left_frame.pack_propagate(False)  # Maintain fixed width
    
    # Right column for input
    right_frame = tk.Frame(main_frame)
    right_frame.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)
    
    # Title for left column
    transactions_title = tk.Label(left_frame, text="Transactions", font=("Arial", 14, "bold"), bg="lightgray")
    transactions_title.pack(pady=10)
    
    # Scrollable frame for transactions list
    canvas = tk.Canvas(left_frame, bg="lightgray")
    scrollbar = tk.Scrollbar(left_frame, orient="vertical", command=canvas.yview)
    scrollable_frame = tk.Frame(canvas, bg="lightgray")
    
    scrollable_frame.bind(
        "<Configure>",
        lambda e: canvas.configure(scrollregion=canvas.bbox("all"))
    )
    
    canvas.create_window((0, 0), window=scrollable_frame, anchor="nw")
    canvas.configure(yscrollcommand=scrollbar.set)
    
    canvas.pack(side="left", fill="both", expand=True)
    scrollbar.pack(side="right", fill="y")
    
    # Right side content
    title_label = tk.Label(right_frame, text="Financial Tracker", font=("Arial", 16, "bold"))
    title_label.pack(pady=20)
    
    # Name input section
    name_label = tk.Label(right_frame, text="Transaction Name:")
    name_label.pack(pady=(10, 5))
    
    name_entry = tk.Entry(right_frame, width=40, font=("Arial", 12))
    name_entry.pack(pady=5)
    
    # Cost input section  
    cost_label = tk.Label(right_frame, text="Cost (£):")
    cost_label.pack(pady=(10, 5))
    
    cost_entry = tk.Entry(right_frame, width=40, font=("Arial", 12))
    cost_entry.pack(pady=5)
    
    # Date input section
    date_label = tk.Label(right_frame, text="Date:")
    date_label.pack(pady=(10, 5))
    
    date_entry = tk.Entry(right_frame, width=40, font=("Arial", 12))
    date_entry.pack(pady=5)
    
    # Button to get the input and store it
    def get_input():
        name = name_entry.get()
        cost = cost_entry.get()
        date = date_entry.get()
        
        if name and cost and date:  # Only add if all fields have content
            # Create a frame for this transaction
            transaction_frame = tk.Frame(scrollable_frame, bg="white", relief=tk.RIDGE, bd=1)
            transaction_frame.pack(fill=tk.X, padx=5, pady=2)
            
            # Add the transaction details
            name_label = tk.Label(transaction_frame, text=f"Name: {name}", bg="white", anchor="w")
            name_label.pack(fill=tk.X, padx=5, pady=2)
            
            cost_label = tk.Label(transaction_frame, text=f"Cost: £{cost}", bg="white", anchor="w")
            cost_label.pack(fill=tk.X, padx=5, pady=2)
            
            date_label = tk.Label(transaction_frame, text=f"Date: {date}", bg="white", anchor="w")
            date_label.pack(fill=tk.X, padx=5, pady=2)
            
            # Clear the input fields
            name_entry.delete(0, tk.END)
            cost_entry.delete(0, tk.END)
            date_entry.delete(0, tk.END)
            
            print(f"Added Transaction - Name: {name}, Cost: £{cost}, Date: {date}")
    
    submit_button = tk.Button(right_frame, text="Add Transaction", command=get_input)
    submit_button.pack(pady=20)
    
    # Start the GUI event loop
    window.mainloop()

# Call the function to open the window
open_window()
```

```python
import tkinter as tk

def open_window():
    """Function to create and open a new window"""
    # Create the main window
    window = tk.Tk()
    
    # Set window properties
    window.title("Financial Tracker")
    window.geometry("800x600+300+100")  # width x height + x_offset + y_offset
    window.minsize(600, 400)
    
    # Create main frame to hold left and right sections
    main_frame = tk.Frame(window)
    main_frame.pack(fill=tk.BOTH, expand=True, padx=20, pady=20)
    
    # Left column for stored transactions
    left_frame = tk.Frame(main_frame, bg="lightgray", width=300)
    left_frame.pack(side=tk.LEFT, fill=tk.BOTH, expand=False, padx=(0, 10))
    left_frame.pack_propagate(False)  # Maintain fixed width
    
    # Right column for input
    right_frame = tk.Frame(main_frame)
    right_frame.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)
    
    # Title for left column
    transactions_title = tk.Label(left_frame, text="Transactions", font=("Arial", 14, "bold"), bg="lightgray")
    transactions_title.pack(pady=10)
    
    # Scrollable frame for transactions list
    canvas = tk.Canvas(left_frame, bg="lightgray")
    scrollbar = tk.Scrollbar(left_frame, orient="vertical", command=canvas.yview)
    scrollable_frame = tk.Frame(canvas, bg="lightgray")
    
    scrollable_frame.bind(
        "<Configure>",
        lambda e: canvas.configure(scrollregion=canvas.bbox("all"))
    )
    
    canvas.create_window((0, 0), window=scrollable_frame, anchor="nw")
    canvas.configure(yscrollcommand=scrollbar.set)
    
    canvas.pack(side="left", fill="both", expand=True)
    scrollbar.pack(side="right", fill="y")
    
    # Right side content
    title_label = tk.Label(right_frame, text="Financial Tracker", font=("Arial", 16, "bold"))
    title_label.pack(pady=20)
    
    # Name input section
    name_label = tk.Label(right_frame, text="Transaction Name:")
    name_label.pack(pady=(10, 5))
    
    name_entry = tk.Entry(right_frame, width=40, font=("Arial", 12))
    name_entry.pack(pady=5)
    
    # Cost input section  
    cost_label = tk.Label(right_frame, text="Cost (£):")
    cost_label.pack(pady=(10, 5))
    
    cost_entry = tk.Entry(right_frame, width=40, font=("Arial", 12))
    cost_entry.pack(pady=5)
    
    # Error message label for cost validation
    cost_error_label = tk.Label(right_frame, text="", fg="red", font=("Arial", 10))
    cost_error_label.pack()
    
    # Date input section
    date_label = tk.Label(right_frame, text="Date:")
    date_label.pack(pady=(10, 5))
    
    date_entry = tk.Entry(right_frame, width=40, font=("Arial", 12))
    date_entry.pack(pady=5)
    
    # Button to get the input and store it
    def get_input():
        name = name_entry.get()
        cost = cost_entry.get()
        date = date_entry.get()
        
        # Clear any previous error message
        cost_error_label.config(text="")
        
        # Validate cost is a float
        try:
            cost_float = float(cost)
            if cost_float < 0:
                cost_error_label.config(text="Cost cannot be negative")
                return
        except ValueError:
            cost_error_label.config(text="Please enter a valid number for cost")
            return
        
        if name and cost and date:  # Only add if all fields have content
            # Create a frame for this transaction
            transaction_frame = tk.Frame(scrollable_frame, bg="white", relief=tk.RIDGE, bd=1)
            transaction_frame.pack(fill=tk.X, padx=5, pady=2)
            
            # Add the transaction details
            name_label = tk.Label(transaction_frame, text=f"Name: {name}", bg="white", anchor="w")
            name_label.pack(fill=tk.X, padx=5, pady=2)
            
            cost_label = tk.Label(transaction_frame, text=f"Cost: £{cost_float:.2f}", bg="white", anchor="w")
            cost_label.pack(fill=tk.X, padx=5, pady=2)
            
            date_label = tk.Label(transaction_frame, text=f"Date: {date}", bg="white", anchor="w")
            date_label.pack(fill=tk.X, padx=5, pady=2)
            
            # Clear the input fields
            name_entry.delete(0, tk.END)
            cost_entry.delete(0, tk.END)
            date_entry.delete(0, tk.END)
            
            print(f"Added Transaction - Name: {name}, Cost: £{cost_float:.2f}, Date: {date}")
    
    submit_button = tk.Button(right_frame, text="Add Transaction", command=get_input)
    submit_button.pack(pady=20)
    
    # Start the GUI event loop
    window.mainloop()

# Call the function to open the window
open_window()
```
